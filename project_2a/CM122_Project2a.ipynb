```python
import sys
import argparse
import numpy as np
import time
import zipfile
import math
import textwrap
import pprint as pprint

from collections import defaultdict, Counter, OrderedDict

test_raw = '/content/test.fasta'
bound_raw = '/content/bound.fasta'
notbound_raw = '/content/notbound.fasta'
pwm = "/content/project2a_PWM.txt"

# build array of sequences -> later, reconstruct seq name by just returning index
def file(fname, content_type):

  # new array element for each newline
  contents = []
  with open(fname, "r") as fh:
    for line in fh:
      contents.append(line) 

  if content_type == 'f': #fasta file
    return [''.join([segment[:-1] for segment in contents[5 * i + 1:5 * i + 5]]) for i in range(len(contents) // 5)]
 
  elif content_type == 't':
    # return dictionary
    dic = {}
    # contents should have 4 rows
    print(contents)
    ACGT = ['A','C','G','T']
    for x in range(0, len(ACGT)):
      # for i in range(0,22): # i manually counted :(
      arr = contents[x].split('\t')
      for element in range(len(arr)): arr[element] = float(arr[element])
      dic[ACGT[x]] = arr
    return dic

def get_file_data(bound_file, notbound_file, test_file):
  return file(bound_file, 'f'), file(notbound_file, 'f'), file(test_file, 'f'), file(pwm, 't')
```

```python
len_pwm = 21 #length of pwm = length of bound transcription factor
pwm_processed = file(pwm, 't') 

def score(test_string):
  # score each index of test string against each index of pwm
  score = 1
  for i in range(len(test_string)): 
    score *= pwm_processed[test_string[i]][i] # value
  return score

def iterate():
  test_processed = file(test_raw, 'f')
  total_scores = [['seq' + str(x + 1), 0] for x in range(len(test_processed))]
  for k in range(len(test_processed)):
    max_score = 0
    for i in range(len(test_processed[k]) - len_pwm + 1):
      curr_score = score(test_processed[k][i:i+len_pwm])
      if curr_score > max_score: max_score = curr_score
    total_scores[k][1] = max_score
  return total_scores
```

```python
# MAIN FUNCTION: output top 2000
def output2000():
  total_scores = iterate()
  total_scores_sorted = sorted(total_scores, key = lambda x: x[1], reverse = True)
  # return(total_scores_sorted[0:1999])
  for i in range(0,2000): print(total_scores_sorted[i][0])
  pass

output2000()
# we don't need name, we can just reconstruct that from index
```
